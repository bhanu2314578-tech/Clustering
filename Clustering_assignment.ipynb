1. What is unsupervised learning in the context of machine learning

    Ans:- Unsupervised Learning is a type of machine learning where the model learns patterns from unlabeled data. The algorithm tries to discover hidden structures, clusters or relationships without predefined outputs. Examples include clustering, dimensionality reduction and anomaly detection.

2. How does K-Means clustering algorithm work?

    Ans:- K-Means works by:
      - Selecting K initial cluster centroids.
      - Assigning each data point to the nearest centroid.
      - Updating centroids as the mean of assigned points.
      - Repeating steps until centroids stabilize.

    The goal is to minimize within cluster variance (inertia)

3. Explain the concept of a dendrogram in hierarchical clustering.

    Ans:- A dendrogram is a tree-like diagram that shows how clusters are merged or split at different distances. It helps visualize cluster hierarchy and determine the optimal number of clusters.

4. What is the main difference between K-Means and Hierarchical Clustering

    Ans:-

| K-Means                                | Hierarchical                   |
| -------------------------------------- | ------------------------------ |
| Requires number of clusters beforehand | No need to predefine clusters  |
| Fast for large datasets                | Computationally expensive      |
| Produces flat clusters                 | Produces hierarchy of clusters |

5. What are the advantages of DBSCAN over K-Means?

    Ans:-
          - Detects clusters of arbitrary shapes.
          - Handles noise/outliers effectively.
          - Does not require specifying number of clusters.
          - Works well for uneven cluster sizes.

6. When would you use Silhouette Score in clustering

    Ans:- Silhouette score is used to evaluate clustering quality by measuring how similiar a point is to its own cluster compared to other clusters. It helps choose the optimal number of clusters.

7. What are the limitations of Hierarchical Clustering

    Ans:-
      - High computational cost.
      - Difficult to scale to large datasets.
      - Sensitive to noise and outliers.
      - Once merged, clusters cannot be split.

8. Why is feature scaling important in clustering algorithms like K-Means

    Ans:- K-means relies on distance calculations. Features with larger scales dominate distance metrics, leading to biased clusters. Scaling ensures all features contribute equally.

9. How does DBSCAN identify noise points?

    Ans:- DBSCAN classifies points as noise if they do not belong to any dense region, meaning they have fewer neighbors than the minimum threshold within a given radius (eps).

10. Define inertia in the context of K-Means.

    Ans:- Inertia is the sum of squared distances between data points and their assigned cluster centroid. Lower inertia indicates more compact clusters.
    
11. What is the elbow method in K-Means clustering

    Ans:- The elbow method helps find the optimal number of clusters by plotting inertia against the number of clusters and selecting the point where the decrease in inertia slows significantly.

12. Describe the concept of "density" in DBSCAN.

    Ans:- Density refers to the number of points within a neighborhood defined by radius (eps). Regions with sufficient density form clusters, while sparse regions become noise.

13. Can hierarchical clustering be used on categorical data

    Ans:- Yes, hierarchical clustering can be applied to categorical data if an appropriate distance metric is used.

14. What does a negative Silhouette Score indicate

    Ans:- A negative score indicates that points may be assigned to the wrong cluster because they are closer to another cluster than their own.

15. Explain the term "linkage criteria" in hierarchical clustering.

    Ans:- Linkage criteria determine how distances between clusters are calculated:-

    - Single linkage
    - Complete linkage
    - Average linkage
    - Ward linkage

16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

    Ans:- K-Means assumes clusters are spherical and equally sized. It struggles with irregulars shapes, different densities and overlapping clusters.

17. What are the core parameters in DBSCAN, and how do they influence clustering?

    Ans:-
    - eps: Neighborhood radius
    - min_samples: Minimum points required to form a dense region

    These parameters determine cluster density and noise detection.

18. How does K-Means++ improve upon standard K-Means initialization?

    Ans:- K-Means++ selects initial centroids more strategically, spreading them apart to reduce poor initialization and improve convergence.

19. What is agglomerative clustering?

    Ans:- Agglomerative clustering is a bottom-up hierarchical clustering method where each data point starts as its own cluster and clusters are merged step-by-step.

20. What makes Silhouette Score a better metric than just inertia for model evaluation?

    Ans:- Silhouette Score considers both:
      - Cluster cohesion (within-cluster similarity)
      - Cluster separation (distance between clusters)


21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a
scatter plot.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X, _ = make_blobs(n_samples=300, centers=4, random_state=1 )
kmeans = KMeans(n_clusters=4)
labels = kmeans.fit_predict(X)
plt.title("K-Means Scatter")
plt.scatter(X[:,0], X[:,1], c=labels)
plt.show()
```

29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.

```python
from sklearn.metrics import silhouette_score
X, _ = make_blobs(n_samples=300, centers=4, random_state=1)

scores = []

for k in range(2,6):
  kmeans = KMeans(n_clusters=k, ).fit_predict(X)
  scores.append(silhouette_score(X, kmeans))

plt.bar(range(2,6), scores)
plt.show()
```

31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with
decision boundaries.

```python
import numpy as np
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=2.5, random_state=5)
model  = KMeans(n_clusters=3).fit(X)

xx, yy = np.meshgrid(
    np.linspace(X[:,0].min()-1, X[:,1].max()+1, 200),
    np.linspace(X[:,0].min()-1, X[:,1].max()+1, 200)
)

Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z  = Z.reshape(xx.shape)
plt.contourf(xx,yy, Z, alpha=0.3)
plt.scatter(X[:,0], X[:,1], c=model.labels_)
plt.show()
```
